```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import log as ln, cumsum
from scipy.integrate import simpson

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
x0 = 0.00001
x1 = 1.
N = 1000
dx = (x1 - x0) / N
x = np.linspace(x0, x1, N)

v0 = 5.
xr = .8
Dc = v0**2/16


for Dt in np.linspace(.2, 1.2, 8)*Dc:
    v = lambda r : v0*(1 - r/xr)
    f = lambda r : r * (ln(r) - 1) + np.cumsum( 1/2*ln( v(r)**2 + 2*Dt ) ) * dx

    plt.plot(x, f(x),'k')
```
