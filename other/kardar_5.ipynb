```python
import numpy as np
import matplotlib.pyplot as plt

from numpy import exp, pi

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
k = (4*pi)/2 # eps/C
# k = 2
u = 0
v = 0
n = 2

A = np.array([
    [k - 4*(n+8)*u, -24*u],
    [-48*v, k-12*(4*u+3*v)]
])

def f(u, v):
    dudl = k*u - 4*((n+8)*u**2 + 6*u*v)
    dvdl = k*v - 4*(12*u*v + 9*v**2)
    return np.array([dudl, dvdl])

fx = ((0, 0), (0, k/36), (k/(n+8)/4, 0), (k/(12*n), k*(n-4)/(36*n)))
print(fx)
```

```python
N = 200
a = .6
u = np.linspace(-.1, .3, N)
v = np.linspace(-.2, .2, N)

u, v = np.meshgrid(u, v)
du, dv = f(u, v)

fig, ax = plt.subplots(figsize=(8, 8))

ax.streamplot(u, v, du, dv, density=2., linewidth=1)
ax.set_ylim(-.2, .2)

for p in fx:
    ax.plot(*p, 'ro')

plt.show()
```
