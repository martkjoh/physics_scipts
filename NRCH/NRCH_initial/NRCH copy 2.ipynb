```python
import numpy as np
from numpy.linalg import eigvals
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.colors import LogNorm
from matplotlib.animation import FuncAnimation as FA

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
c11 = .2
c12 = .5
c21 = .1
c22 = .5
x = -.2
xx = .2

def f(x1, x2, a):
    mu1 = -2*(x1 - c11)*(x1 - c12)*(2*x1 - c11 - c12) + (x+a)*x2 + 2*xx*x1*x2**2
    mu2 = -2*(x2 - c21)*(x2 - c22)*(2*x2 - c21 - c22) + (x-a)*x1 + 2*xx*x2*x1**2
    return np.array([mu1, mu2])
```

```python
N = 100
a = .9
x1, x2 = np.linspace(-a, a, N), np.linspace(-a, a, N)
x1, x2 = np.meshgrid(x1, x2)

aa = np.linspace(0, 3., 9)
fig, ax = plt.subplots(3, 3, figsize=(30, 30))

for n, a in enumerate(aa):
    u, v = f(x1, x2, a )
    l = np.sqrt( u**2 + v**2)
    ax[n//3, n%3].streamplot(x1, x2, u, v, color=l, density=1., linewidth=.6)
    ax[n//3, n%3].set_title("$\\alpha={}$".format(a))
```
