```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.animation import FuncAnimation as FA

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
def l(q, k, a):
    k = k+0j 
    a = a+0j
    q = q+0j
    return (
    k**4 + (2*1j)*a*k*q - q**2 - q**4 - k**2*(1 + 4*q**2) + np.sqrt(k**8 - (-1 + a**2)*q**4 - 2*k**6*(1 + q**2) - (4*1j)*a*k*q**3*(1 + 2*q**2) + k**4*(1 + 4*q**2 + q**4) + 2*k**2*(q**2 + 7*q**4 + 8*q**6))
    )

def le(q, k, a):
    k = k+0j 
    a = a+0j
    q = q+0j
    return (
        q*((2*1j)*a*k - ((2*1j)*a*q**2)/(k - k**3) + (-1 + a**2/(2*k**2*(-1 + k**2)) - (2*(2 - 7*k**2 + 4*k**4))/(-1 + k**2)**3)*q**3 + (k**2*q - 3*k**4*q)/(-1 + k**2))
        )
```

```python
from matplotlib.lines import Line2D

q = np.linspace(0, 0.6, 500)

fig, ax = plt.subplots(1, 3, figsize=(18, 3), sharey=True, sharex=True)

kk = [.4, .5, .6]
for i, k in enumerate(kk):

    ax[i].ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    ax[i].plot(q, 0*q, 'k--', lw=1)
    ax[i].set_xlabel("$q$")

    aa = [0.5, 1, 1.5, 2]
    for j, a in enumerate(aa):
        ax[i].plot(q, np.real(l(q, k, a)), color=cm.viridis(j/(len(aa)-1)))
        ax[i].plot(q, np.real(le(q, k, a)), '--', color=cm.viridis(j/(len(aa)-1)))

    ax[i].set_title("$k = {k:.1f}$".format(k=k))
ax[0].set_ylim(-6.5e-3, 9e-4)
ax[0].set_xlim(0,0.6)
ax[0].set_ylabel("$\\mathrm{Re}\\,\\lambda_+(q)$")
fig.subplots_adjust(wspace=0)



sm = plt.cm.ScalarMappable(cmap=cm.viridis, norm=plt.Normalize(vmin=min(aa), vmax=max(aa)))
sm.set_array([])
cbar = fig.colorbar(sm, ax=ax, orientation='vertical', fraction=0.02, pad=0.04)
cbar.set_label('$\\alpha$')


legend_elements = [
    Line2D([0], [0], color='black', lw=2, linestyle='-', label='Full expession'),
    Line2D([0], [0], color='black', lw=2, linestyle='--', label='Expanded $\\mathcal{O} (q^4)$')
]
ax[-1].legend(handles=legend_elements, loc='lower right', fontsize=12)


fig.subplots_adjust(bottom=0.25, right=0.85)
plt.savefig("fig/eigen.pdf")
```
