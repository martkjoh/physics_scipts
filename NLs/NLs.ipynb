```python
import numpy as np
from numpy.linalg import eigvals
import matplotlib.pyplot as plt

from matplotlib import cm
from matplotlib.colors import LogNorm, Normalize

plt.style.use("../rc")
```

```python
d = 1

def f(g, t):
    bt = -(d-2)*t + (d-1)/(2*d)*g*t
    bg = -(d-2)*g + (2*d-3)/(2*d)*g**2 - (7*d - 4)/d*g*t
    return np.array([bg, bt])

# e = d - 2
# def f(g, t):
#     bt = -e*t + 1/4*g*t
#     bg = -e*g + 1/4*g**2 - 5*g*t
#     return np.array([bg, bt])
```

```python
N = 100
k1 = 6
k2 = .8
h = 0.1
steps = 50
maxlen = np.sqrt(k1**2+k2**2)

x1, x2 = np.linspace(0, k1, N), np.linspace(0, k2, N)
x1, x2 = np.meshgrid(x1, x2)

u, v = f(x1, x2)
colors = np.zeros_like(x1) 

for ii in range(x1.shape[0]):
    for jj in range(x1.shape[1]):
        g = x1[ii, jj]
        t = x2[ii, jj]
        
        for _ in range(steps):            
            dg, dt = f(g, t)
            g += h * dg 
            t += h * dt
            # if t > 1: t = maxlen; break

        c = np.sqrt(g**2 + t**2)
        # c = np.abs(g)

        if np.isnan(c): c = maxlen
        colors[ii, jj] = min(c, maxlen)
```

```python
size = 4
fig, ax = plt.subplots(figsize=(size + 2, size), sharex=True, sharey=True)

u, v = f(x1, x2)
ax.streamplot(x1, x2, u, v, density=2., linewidth=.6, color='black')
ax.set_title("$d={:.0f}$".format(d))
ax.set_xlabel("$g$")

ax.pcolormesh(
    x1, x2, colors,
    cmap=cm.coolwarm,
    alpha=.7,
    linewidths=0.1,
    rasterized=True
)

ax.set_ylabel("$\\theta$")
ax.set_ylim(0,k2)
ax.set_xlim(0,k1)

plt.savefig("d={}.svg".format(d))
```

### Arbitrary $\pi$ renorm

```python
d = 3
Bs = [-1, 0, 0.1, 0.25, .5, 1]

N = 50
k1 = 8
k2 = .6
h = 0.2
steps = 50
maxlen = np.sqrt(k1**2+k2**2)

x1, x2 = np.linspace(0, k1, N), np.linspace(0, k2, N)
x1, x2 = np.meshgrid(x1, x2)

def f(g, t, B):
    bt = -(d-2)*t + ((d-1)/(2*d) - B)*g*t
    bg = -(d-2)*g + (2*d-3)/(2*d)*g**2 - (7*d - 4)/d*g*t
    return np.array([bg, bt])

size = 4
fig, axs = plt.subplots(1, len(Bs), figsize=(size * len(Bs) + 2, size ), sharex=True, sharey=True)

for i, B in enumerate(Bs):
    
    u, v = f(x1, x2, B)
    colors = np.zeros_like(x1) 
    for ii in range(x1.shape[0]):
        for jj in range(x1.shape[1]):
            g = x1[ii, jj]
            t = x2[ii, jj]
            
            for _ in range(steps):            
                dg, dt = f(g, t, B)
                g += h * dg 
                t += h * dt
            c = np.sqrt(g**2 + t**2)
            if np.isnan(c): c = maxlen
            colors[ii, jj] = min(c, maxlen)

    ax = axs[i]
    ax.streamplot(x1, x2, u, v, density=2., linewidth=.6, color='black')
    ax.set_title("$B={:.2f}$".format(B))
    ax.set_xlabel("$g$")
    ax.pcolormesh(
        x1, x2, colors,
        cmap=cm.coolwarm,
        alpha=.7,
        linewidths=0.1,
        rasterized=True
    )

axs[0].set_ylabel("$\\theta$",y=1.04)
ax.set_ylim(0,k2)
ax.set_xlim(0,k1)

fig.suptitle("$d={:.0f}$".format(d))
plt.show()
```

```python
n = 100
k1 = 5
k2 = .7
x1, x2 = np.linspace(0, k1, n), np.linspace(0, k2, n)
x1, x2 = np.meshgrid(x1, x2)
u, v = f(x1, x2)

colors = np.zeros_like(x1) 
h = 0.1
steps = 100
maxlen = 5

for ii in range(x1.shape[0]):
    for jj in range(x1.shape[1]):
        g = x1[ii, jj]
        t = x2[ii, jj]
        
        for _ in range(steps):
            dg, dt = fN(g, t)
            g += h * dg 
            t += h * dt

        c = np.sqrt(g**2 + t**2)
        c = np.abs(g)
        if np.isnan(c): c = maxlen
        colors[ii, jj] = min(c, maxlen)
```
