```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import cm
from numpy import exp, heaviside, log as ln
th = lambda x : heaviside(x, 1/2)

plt.rc("font", family="serif", size=16)
plt.rc("mathtext", fontset="cm")
plt.rc("lines", lw=2)
```

```python
x = np.linspace(0.01, 4.05, 10000)
r1 = lambda y: 1 / (exp(y) - 1)
r2 = lambda y : (1 - y) / y * th(1 - y)

dy = x[1]**2 - x[0]**2
D =  lambda r, y : (r(y + dy/2) - r(y - dy/2)) / dy

fig, ax = plt.subplots(1,2, figsize=(16, 6))
ax[0].plot(x, r1(x**2), 'k', label=r"$r_1(y) = \frac{1}{ e^y - 1 }$")
ax[0].plot(x, r2(x**2), 'r', label=r"$r_2(y) = \frac{1 - y}{y}\theta(1 - y) $")

ax[0].set_xlabel("$q / k$")
ax[0].set_ylabel("$r(q^2 / k^2)$")
ax[0].set_ylim(-.1, 1.9)

ax[1].plot(x, x**2 *  r1(x**2), 'k', label=r"$R_{1,k}(q) = q^2 r_1(q^2/k^2) $")
ax[1].plot(x, x**2 *  r2(x**2), 'r', label=r"$R_{2,k}(q) = q^2 r_2(q^2/k^2) $")
ax[1].set_xlabel("$q / k$")
ax[1].set_ylabel("$R_k(q)$")

ax2 = ax[1].twinx()
ax2.plot(x, - 2 * x**4 * D(r1, x**2), 'k--', label=r"$k\partial_k R_{1,k}(q)$")
ax2.plot(x, - 2 * x**4 * D(r2, x**2), 'r--', label=r"$k\partial_k R_{2,k}(q)$")
# ax2.set_ylim(-.1, 1.9)


ax[0].legend()
ax[1].legend()
ax2.legend(loc=5)
fig.savefig("fig/regulators.pdf")
```
